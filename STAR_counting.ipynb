```
from matplotlib import pyplot as plt
from skimage import data 
from skimage.feature import blob_dog, blob_doh, blob_log   # binary large object(BLOB) Laplacian of Gaussian (LoG) Difference of Gaussian (DoG) Determinant of Hessian (DoH)
from math import sqrt
from skimage.color import rgb2gray
from skimage.io import imread
import glob
```

```
path_file = '/stars_count.jpg'
image = imread(path_file, as_gray=True)

plt.imshow(image, cmap='gray')
plt.show()
```

```
blobs_log = blob_log(image, max_sigma=20, num_sigma=10, threshold=.1) #star identifying command
#results in three outputs co-ordinates and Area of objects using standard deviation
#find the radius of each object
blobs_log[:,2]=blobs_log[:,2]* sqrt(2)
total_number_of_objects = len(blobs_log)
print("Total number of Stars in sky are =",total_number_of_objects)
```

To check count it is correct #visualisation of objects identified

```
fig, ax = plt.subplots(1,1)
plt.imshow(image, cmap='gray')
for star in blobs_log:
  y,x,r = star
  c = plt.Circle((x,y), r+5, color='lime', linewidth=2, fill=False)
  ax.add_patch(c)
```
